```python
import geopandas as gpd
import pandas as pd
from libpysal.weights import fuzzy_contiguity

import sys
sys.path.append('../')
from create_yearMonthID import create_yearMonthID

from pytictoc import TicToc
t = TicToc()
```

```python
# Shapefile of polygons
pol_fp = '../../data/testing_buffers_spatial/MBfogo_c10_cell_32deg_polygons_2019.shp'
```

```python
pol = gpd.read_file(pol_fp)
```

```python
# Converting to metre CRS
print("Converting polygon's CRS")
pol = pol.to_crs('epsg:5880')
```

```python
W.component_labels[len(W.component_labels)-1]
```

```python
pols_idd = []
last_W = -1
for month in range(1, 12+1):
    t.tic()
    print('Identifying connected components for month {}'.format(month))

    # Subsetting polygons mapped on month
    pol_month = pol[pol['month'] == month]

    # Identifying the polygons that are within a distance from one another
    W = fuzzy_contiguity(pol_month, buffering = True, buffer = 30, silence_warnings = True)
    
    # Correcting to accumulate index through months
    connected_ids = W.component_labels + last_W + 1
    
    pol_month['connected_ids'] = connected_ids
    
    pols_idd.append(pol_month)
    
    # Updating the last weight position
    last_W = connected_ids[len(connected_ids)-1]
    
    del pol_month, W, connected_ids
    
    t.toc('Connected components for month {} identified in'.format(month))
    
pols_idd = gpd.GeoDataFrame(pd.concat(pols_idd))
```

```python
# On the one hand, we create the CSV lookup table
lookup = pols_idd[['ID','year','month','connected_ids']]
```

```python
# The, also, create a unique ID for each new ID
new_ids = lookup.drop(columns = ['ID']).drop_duplicates(subset = 'connected_ids').reset_index(drop = True)
new_ids
```

```python
new_ids = create_yearMonthID(new_ids).reset_index()
new_ids = new_ids.rename(columns = {'ID': 'ID_sjoin'})
```

```python
new_ids
```

```python
# THis new datafrme, we merge it back to lookup
lookup = lookup.merge(new_ids[['ID_sjoin','connected_ids']], on = ['connected_ids'], how = 'left')
lookup
```

```python
# Now we could drop the connected_ids column. We already have the lookup we wanted that relates the original polygons to the new ones.
```

```python
# Then, on the other hand, we carry out spatial joins. 
# I think, we first join the new ids back to the shapefile, then we do a spatial join based on the ID_sjoin
pols_idd = pols_idd.merge(lookup, how = 'left')
pols_idd.head(40)
```

```python
# Now, spatial join
t.tic()
new_pols = pols_idd[['ID_sjoin','year','month','geometry']].dissolve(by = 'ID_sjoin').reset_index()
t.toc()
```
